```python

driver = webdriver.Chrome('D:\ScrapingImdbData\chromedriver_win32\chromedriver.exe') #connecting to chrom driver
driver.get("https://www.imdb.com/list/ls041677842/")  
response = driver.page_source   
soup = BeautifulSoup(response)  
 
movie_details=[]
movie_summary=[]
movie_rating=[]
movie_tagline=[]
movie_director=[]
movie_genre=[]
movie_budget=[]
movie_gross=[]
movie_writer=[]
movie_stars=[]
movie_name2=[]
movie_name3=[]
movie_name4=[] 
ids=[]
idx=0
PID_Details=[]
PID_Details2=[]
PID_Details3=[]
ID_Genre=[]
j=0
ID_Director=[]
p=0
ID_Writer=[]
q=0
i=0
r=0
ID_Stars=[]
PID_Details4=[]
movie_name5=[]

print("Searching2019....")
for link in soup.findAll('h3', attrs={'class':'lister-item-header'}):  
    driver.get('https://www.imdb.com'+link.a['href']) 
    
    idx=idx+1
    ids.append(idx)
    time.sleep(3)
    contentnew=driver.page_source
    datanew=BeautifulSoup(contentnew)
    #scraping the title of movie from new page
    summary=datanew.find('div',attrs={'class':'summary_text'})
    movie_summary.append(summary.text.strip())
    #print(movie_summary)
    rating=datanew.find('div',attrs={'class':'ratingValue'})
  
    #tagline=datanew.find('div', attrs={'class':'txt-block'})
    #print(tagline.text)
    #movie_tagline.append(tagline.text.strip())
    title=datanew.find('div', attrs={'class':'title_wrapper'})
    #print(title.h1.text) # checking title 
    movie_rating.append(rating.text.strip())
    #print(rating.text)
    movie_details.append(title.h1.text) # adding to array
    for summary in datanew.findAll('div', attrs={'class':'txt-block'}):
            if(summary.h4==None):
                movie_tagline.append("N/A")
                #print("N/A")
                break;
            if(summary.h4.text=='Taglines:' or summary.h4.text=='Tagline:'):
                    movie_tagline.append(summary.text.strip().replace("Taglines:","").strip())
                    #print(summary.text)
                    #print(movie_tagline)
                    break;
    for summary in datanew.findAll('div', attrs={'class':'see-more inline canwrap'}):
            if(summary.h4==None):
                movie_tagline.append("N/A")
                #print("N/A")
                break;
            if(summary.h4.text=='Genres:' or summary.h4.text=='Genres:'  ):
                for d in summary.findAll('a'):
                    movie_genre.append(d.text)
                    j=j+1
                    ID_Genre.append(j)
                    PID_Details.append(i+1)
                    movie_name2.append(title.h1.text)
                    #print(summary.text)
                break
    for summary in datanew.findAll('div', attrs={'class':'txt-block'}):
            if(summary.h4==None):
                movie_budget.append("N/A")
                #print("N/A")
                break;
            if(summary.h4.text=='Budget:' or summary.h4.text=='Budget:'  ):
                #print("asd")
                x=summary.text.replace("Budget:","")
                movie_budget.append(x.strip().replace("(estimated)","").strip())
                #print(summary.text)
                break
    for summary in datanew.findAll('div', attrs={'class':'txt-block'}):
            if(summary.h4==None):
                movie_gross.append("N/A")
                #print("N/A")
                break;
            if(summary.h4.text=='Cumulative Worldwide Gross:' or summary.h4.text=='Cumulative Worldwide Gross:'  ):
                x=summary.text.replace("Cumulative Worldwide Gross:","")
                movie_gross.append(x.strip().replace("(estimated)","").strip())
                #print(summary.text)
                break
    chk=0
    for summary in datanew.findAll('div', attrs={'class':'credit_summary_item'}):
            if(summary.h4==None):
                #movie_stars.append("N/A")
                #print("N/A")
                break;
            if(summary.h4.text=='Stars:' or summary.h4.text=='Stars:'  ):
                #print("asd")
                for d in summary.findAll('a'):
                    y=d.text.replace("See full cast & crew","")
                    movie_stars.append(y.strip().replace("Stars:","").strip())
                    r=r+1
                    ID_Stars.append(r)
                    PID_Details4.append(i+1)
                    movie_name5.append(title.h1.text)
                    #print(summary.text)
                    chk=chk+1
                break
    if(chk==0):
            movie_stars.append("N/A")
            r=r+1
            ID_Stars.append(r)
            PID_Details4.append(i+1)
            movie_name5.append(title.h1.text)
    for summary in datanew.findAll('div', attrs={'class':'credit_summary_item'}):
            if(summary.h4==None):
                movie_writer.append("N/A")
                #print("N/A")
                break;
            if(summary.h4.text=='Writers:' or summary.h4.text=='Writer:'  ):
                #print("asd")
                for d in summary.findAll('a'):
                    movie_writer.append(d.text.strip().replace("Writers:","").strip())
                    q=q+1
                    ID_Writer.append(q)
                    PID_Details3.append(i+1)
                    movie_name4.append(title.h1.text)

                    #print(summary.text)
                break
                    
    for summary in datanew.findAll('div', attrs={'class':'credit_summary_item'}):
        if(summary.h4.text=='Directors:' or summary.h4.text=='Director:'  ):
            for d in summary.findAll('a'):
                x=d.text.strip().replace("Directors:","")
                movie_director.append(x.strip().replace("Director:","").strip())
                p=p+1
                ID_Director.append(p)
                PID_Details2.append(i+1)
                movie_name3.append(title.h1.text)
                #print(movie_director)
            break;


   
    
    
    
    i=i+1
    


 
print("-------------------------")
print(movie_details)



```
